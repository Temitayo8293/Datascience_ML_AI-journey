1. Create a NumPy array of 50 random integers between 1 and 100.
    * Compute:
      - Mean
      - Median
      - Standard deviation
    * Find:
        - Maximum value
        - Minimum value

```python
import numpy as np
import pandas as pd
```

```python
random_array_number = np.random.randint(1, 100, size = 50)
print (f'Random array number = {random_array_number}') # This print the generated random array
print ('')
print (f'Results:')
print (f'The mean of the random array is "{np.mean(random_array_number)}"')  #This print the mean of the array
print (f'The median of the random array is "{np.median(random_array_number)}"') # This print the median of the array
print (f'The standard deviation of the random array is "{np.std(random_array_number)}"') # This print the standard deviation of the array
print (f'The minimum value of the random array is "{np.min(random_array_number)}"') # This print the minimum value of the array
print (f'The maximum value of the random array is "{np.max(random_array_number)}"') # This print the maximum value of the array
```

```python
greater_than_60 = random_array_number[random_array_number>60]
less_than_30 = random_array_number[random_array_number<30]

print(f'Numbers greater than 60 in the array are {greater_than_60}') # This print the arrays greater than 60
print(f'Numbers less than 30 in the array are {less_than_30}') # This print the array less than 30
```

```python
#Plot a bar chart of job title vs salary_year_avg#

import matplotlib.pyplot as plt
plt.bar (df_employee['job_title'], df_employee['salary_year_avg'])
plt.ylabel('Salary year average')
plt.title('Bar chart of Job title vs Salary year average')
plt.xticks(rotation =45, ha = 'right')
plt.show()

 
```
