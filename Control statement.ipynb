### Control flow

#### conditional statement

* `if` statement

```python
attendance = 70

if attendance >= 70:
    print ('You are certified')
```

```python
attendance = 50

if attendance >= 70:
    print ('You are certified')
```

* `if-else` statement

```python
attendance = 50

if attendance >= 70:
    print('You are qualified')
else:
    print('You are not qualified')
```

Using the `if-else` statement with the input() function

```python
num = int(input('Enter a number: '))

if num % 2 == 0:
    print('This is an even number')
else:
    print('This is an odd number')
          
          
```

* if-elif-else statement


```python
score = int(input('Enter your score: '))

if score >= 90:
    print('Grade A')
elif score >=70:
    print ('Grade B')
elif score >=50:
    print ('Grade C')
else:
    print ('Grade F')
```

#### Looping statements


* for loop

```python
# perform iterations.

colors = ['red', 'yellow', 'green', 'black']
print(colors)
```

```python
for color in colors:
    print(color)
```

```python
# using range() function inclusive, exclusive
for i in range(1, 10):
    print(i)
```

using for loop for list comprehension

```python
for x in range(5):
    print(x)
```

```python
# using list comprehension

numbers = [y for y in range(5)]
print(numbers)
```

```python
add_num = [x + 2 for x in range(5)]
print(add_num)
```

* Filter elements in list comprehension using an if statement

```python
# filter out odd number in the list
odd_num = [x for x in range(10) if x % 2 != 0]
print (odd_num)
```

```python
# filter out even number
even_num = [y for y in range(10) if y % 2 == 0]
print (even_num)
```

* While loop


```python
# count from 1 to 10

count = 1

while count <=10:
    print(f'Count: {count}')
    count += 1 #incremnt count
```

### loop control flow statement

* using break

```python
# using break
for num in range(1,10):
    if num == 5:
        break #stops loop process
    print(num)
```

* using continue

```python
# using continue
for num in range(1,9):
    if num == 3:
        continue # skips the loop on process 3
    print(num)
```

* Using pass

```python
for num in range(1, 7):
    if num == 3:
        pass #not affecting the program
    else:
        print(num)
```
