```python
#  multiple positional arguments (uses *args)
# used when we need a function that can accept multiple arguments
def sum_numbers(*args):
    addition = sum(args)
    print(f'Total sum = {addition}')

```

```python
sum_numbers(2,3,5)
```

```python
sum_numbers(50,200,400,1000)
```

```python
# multiple keyword arguments (to pass multiple name arguments)
# make use of **kwargs

#example: cretaing student info

def student_info(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}:{value}')
```

```python
student_info(Name = 'Sola', Age = '18', Track = 'AI')
```

```python
# multiple return values

def student_score(name, maths, english, science):
    total = maths + english + science
    avg = total/3
    return name, total, avg
```

```python
student_name, total_score, average = student_score('Paul', 70, 65, 80)
```

```python
print(f'Welcome {student_name}, your total score is {total_score}, and the average is {average}')
```

## Scope in python
* Local vs global scope

```python
# Local scope variables (defined inside a function: variable existing only within the block of code in a function)

def show_data():
    balance = '500mb'# (local scope variable)
    print(f'Your data balance is {balance}')
```

```python
show_data ()
#print(balance) will not work becauase it is in a function
```

```python
balance = '500mb'# (global scope variable)
def check_data():
    print(f'Your data balance is {balance}')
```

```python
check_data()
```

```python
print (balance)
```

###### modifying a global variable inside a function

* use `global` to modify a global variable inside the function

```python
count = 0
def increase():
    global count
    count += 2
    print(f'Count: {count}')
```

```python
increase()
```

```python
print (count) #changes the value of global variable
```

```python
fruits = ['apple', 'orange', 'grape']

def new_fruits():
    global fruits
    fruits[0] = 'banana'
    print(fruits)
```

```python
new_fruits()
```

```python
print(fruits)
```

```python
def add_fruits():
    global fruits
    fruits.append('pineapple')
    print(fruits)
```

```python
add_fruits()
```

```python
print (fruits)
```
