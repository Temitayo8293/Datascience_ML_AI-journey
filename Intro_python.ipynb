## Basic operations in python

* Arithmetic operators

```python
a= 30
b = 20

print (a+b)
print (a-b)
# multi
print (a*b)
# division
print (a/b)
# floor
print (a//b)
# modulus
print (a%b)
#exponent
print (a**b)
```

* Assignment operator

```python
#Assign value
x = 5
print (x)
# add and assign
x +=4
print (x)
#substract and assign
x -= 1
print (x)
# multiply and assign
x *= 2
print (x)
# divide and assign
x /= 2
print (x)
```

* Comparison operators

```python
# Equal to (==)
a = 6
b = 4
print (a==b)
# Not equal to (!=)
print (a!=b)
# Greater than (>)
print (a>b)
# Less than (<)
print (a<b)
# Greater than or equal to (>=)
print (a>=b)
# Less than or equal to (<=)
print (a<=b)
```

* Logical operators

```python
#mostly used for decision making
# 'and' logical oeprator
u = 5
v = 10
print (u > 2 and v < 15)
# 'or' logical operator
print (u > 2 or v > 15)
# 'not'
print (not u>2)
```

# Data structure


* List []

```python
# List: []

num_list = [10, 20, 11, 19, 80]
print(num_list)
print()
print(type(num_list))

# List is ordered and changable (mutable)
```

```python
fruits = ['Apple', 'Cherry', 'Banana', 'Mango']
print(fruits)
print()
print(type(fruits))
```

* Tuple ()

```python
# Tuple ()
num_tuple = (10, 20, 11, 19, 80)
print(num_tuple)
print()
print(type(num_tuple))
# Tuple is ordered and immutable
# follows the order you gave
```

* Set {}

```python
num_set1 = {10, 20, 11, 19, 80}
print(num_set1)
print()
print(type(num_set1))
# Set is unordered and immutable
# Set returns unique values only
num_set2 = {10, 20, 11, 19, 80, 10, 20, 11, 19, 80}
print(num_set2)
```

* Dictionary - key value pairs using a curly brackets{}

```python
# it comes in key value pairs
stud_bio = {'name' : 'Oluwafemi', 'Age' : 20, 'Gender' : 'Male'}
print(stud_bio)
print()
print(type(stud_bio))
```

```python
bio_data = {'name' : 'Oluwafemi', 'Age' : 20, 'Gender' : 'Male'}
print(bio_data)
print()
print(type(bio_data))
```
