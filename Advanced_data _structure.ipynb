* Adding elements to list

```python
# using .append()
colors = ['red', 'yellow']
print (colors)
```

```python
colors.append('guava')
print (colors)
```

```python
#using .insert() - allows to insert in a particular index
name = ['femi', 'segun', 'james']
name.insert(1, 'tolu')
print(name)
```

* Removing elements from a list

```python
print (name)
```

```python
# remove using .remove() (updated immediately)
name.remove('taiwo')
print(name)
```

```python
# .pop() (specifying the index)

name.pop(1)
print(name)
```

```python
# using del variable[index]
del name[0]
print(name)
```

### To modify a tuple, we convert it to a list to edit it

```python
# list to tuple
num_list = [3,4,2,6,7]

print (num_list)
print(type(num_list))
```

```python
#list to tuple using tuple()
num_tuple = tuple(num_list)
print(num_tuple)
print(type(num_tuple))
```

```python
#convert tuple to a list
f_tuple = ('mango', 'apple')
print(type(f_tuple))
```

```python
f_list = list(f_tuple)
print(type(f_list))
print(f_list)
```
